```python
from PIL import Image
import torch

from transformers import AutoModel, AutoImageProcessor

REPO_NAME = "p1atdev/MangaLineExtraction-hf"

model = AutoModel.from_pretrained(REPO_NAME, trust_remote_code=True)
processor = AutoImageProcessor.from_pretrained(REPO_NAME, trust_remote_code=True)

image = Image.open("/content/drive/MyDrive/Colab Notebooks/OIP.jpg")

inputs = processor(image, return_tensors="pt")

with torch.no_grad():
    outputs = model(inputs.pixel_values)

line_image = Image.fromarray(outputs.pixel_values[0].numpy().astype("uint8"), mode="L")
line_image.save("./line_image.png")
```

```python
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt  # Для відображення
from transformers import AutoImageProcessor, Swin2SRForImageSuperResolution

# Завантажуємо модель та процесор
processor = AutoImageProcessor.from_pretrained("caidas/swin2SR-classical-sr-x2-64")
model = Swin2SRForImageSuperResolution.from_pretrained("caidas/swin2SR-classical-sr-x2-64")

# Шлях до зображення в Google Drive
image_path = "/content/drive/MyDrive/Colab Notebooks/OIP.jpg"

# Відкриваємо зображення (без використання requests.get())
image = Image.open(image_path)

# Підготовка зображення для моделі
inputs = processor(image, return_tensors="pt")

# Пропускаємо через модель
with torch.no_grad():
    outputs = model(**inputs)

# Обробка вихідного результату
output = outputs.reconstruction.data.squeeze().float().cpu().clamp_(0, 1).numpy()
output = np.moveaxis(output, source=0, destination=-1)
output = (output * 255.0).round().astype(np.uint8)  # Перетворення float32 → uint8

# Створюємо зображення з вихідних даних
output_image = Image.fromarray(output)

# Відображаємо зображення
plt.imshow(output_image)
plt.axis("off")  # Видаляємо осі
plt.show()

# (Опціонально) Зберігаємо результат у Google Drive
output_image.save("/content/drive/MyDrive/Colab Notebooks/super_res_image.jpg")
```
